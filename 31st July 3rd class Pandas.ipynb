```python
df_sales  = pd.read_csv("sales_data_final.csv")
```

```python
df_sales.head()
```

```python
#creating new column to store data from another existing column
```

```python
def profit_flag(a):
    if a >= 0:
        return "positive"
    else:
        return "negative"
```

```python
profit_flag (-45)
```

```python
df_sales["flag_profit"]= df_sales["profit"].apply(profit_flag)
```

```python
df_sales.head()
```

```python
#creating new column to store length of the customer name
```

```python
df_sales['len_cust_name'] = df_sales['customer_name'].apply(len)
```

```python
df_sales.head()
```

```python
#deriving column: catagorize data from a column into 3 segments, like high , low and medium
```

```python
def quantity_flag(a):
    if a<10 :
        return "low"
    elif a> 10 and a<20:
        return "medium"
    else :
        return "high"
```

```python
df_sales["flag_quantity"]=df_sales["quantity"].apply(quantity_flag)
```

```python
df_sales.head(5)
```

```python
#deriving column based on lemda functiond
#line anonymus function, we can create function in one line
```

```python
df_sales["square_quantity"]=df_sales["quantity"].apply(lambda a:a**2)
```

```python
df_sales
```

```python
df_sales["total_revenue"]=df_sales["quantity"]*df_sales["profit"]
```
